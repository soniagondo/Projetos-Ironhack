## Game Project : IronForca

   
**Regras do jogo:**
   1) O jogador poderá cometer até 7 erros na escolha das letras da palavra da jogada.
   
   2) É permitido escolher apenas letras.
   
   3) Não é permitido escolher a mesma letra mais de uma vez.
   
      
**Sequência do jogo:**

   1) Jogador responde "s" para começar o jogo ou "n" para não jogar.
   
   2) Jogador escolhe o idioma das palavras do jogo. "1" para palavras em Português ou "2" para palavras em Inglês.
   
   3) Jogador seleciona letras para tentar descobrir qual a palavra do jogo.
   
   4) Quando a jogada terminar, o jogador poderá escolher se deseja jogar novamente.
   
   5) O número de jogadas ganhas e perdidas é contabilizado no final da jogada.


```python
lista_forca =[f'''
    ————|
    |
    |
    |
    |
    |
    ————''',
    f'''
    ————|
    |   O
    |
    |
    |
    |
    ————''',
    f'''
    ————|
    |   O
    |   |
    |
    |
    |
    ————''',
    f'''
    ————|
    |   O
    |  /|
    |
    |
    ————''',
    f'''
    ————|
    |   O
    |  /|\ \t
    |
    |
    ————''',
    f'''
    ————|
    |   O
    |  /|\ \t
    |  /   \t
    |
    ————''',
    f'''
    ————|
    |   O
    |  /|\ \t
    |  / \ \t
    |
    ————''',
    f'''
    ————|
    |   O
    |  === \t
    |  /|\ \t
    |  / \ \t
    |
    ————'''
    ]
   
```

```python
import random

# variável para armazenar pontuação do jogador
jogadas_ganhas = 0 
jogadas_perdidas = 0

# listas de palavras para serem adivinhadas
lista_palavra_portugues = ['CORONA','TARTARUGA','PARALELEPIPEDO','URUGUAI','MACARRAO','TELEVISAO','TESOURA','PATINETE','TIJOLO','OFTALMOLOGISTA','TATUAGEM']
lista_palavra_ingles = ['WONDERFUL','BEAUTIFUL','DATAFRAME','GORGEOUS','CLASSROOM','POPCORN','VOCABULARY','ENGLISH','SPANISH','DICTIONARY','HANGMAN','MEMORIZE','SUCESS','CONGRATULATIONS']


# pergunta ao jogador qual a letra escolhida, certificando-se que o jogador escolheu uma letra.
def qual_letra():
    
    '''Função pergunta ao jogador qual a letra escolhida. Faz a verificação se o input corresponde a uma letra.
       Retorna a letra escolhida.'''
    
    letra_escolhida = False
    
    while letra_escolhida == False:
        letra = input("Por favor, escolha uma letra. Letra escolhida é:").upper()
        
        if not('A'<= letra <= 'Z') and not(letra == 'Ç'):
            print("Você só pode escolher letras.")
        elif len(letra)>1:
            print("Você só pode informar uma letra.")
        elif letra in letras_escolhidas:
            print("Letra já escolhida.")
            
        else:
            letra_escolhida = True
            
    return letra

# pergunta inicialmente se o jogador quer jogar
quer_jogar = False
    
while quer_jogar == False:
    jogada = str(input('Você quer jogar (s/n)?')).upper()
        
    if not jogada == 'S' and not jogada == 'N':# condições para certificar que o input do jogador está dentro das opções s ou n
        print("Escolha entre as opções s ou n")      
        quer_jogar = False
    elif jogada == 'S': 
        quer_jogar = True
    elif jogada == 'N':
        quer_jogar = True


while jogada == 'S':
    
    tema_escolhido = int(input("Por favor, escolha 1 para jogadas com palavras em Português e 2 para jogadas em Inglês.\nTema escolhido é:"))
              
    if tema_escolhido == 1:
        
        palavra_jogada = random.choice(lista_palavra_portugues) #palavra da jogada será escolhida aleatoriamente dentro da lista_palavra_portugues
        lista_palavra_portugues.remove(palavra_jogada) #remove a palavra da jogada da lista de palavras para não ser escolhida novamente
                
    elif tema_escolhido == 2:
        
        palavra_jogada = random.choice(lista_palavra_ingles) #palavra da jogada será escolhida aleatoriamente dentro da lista_palavra_ingles
        lista_palavra_ingles.remove(palavra_jogada) #remove a palavra da jogada da lista de palavras para não ser escolhida novamente
             
    # lista a quantidade de espaços que possui a palavra da jogada
    lista_espaços = []
    for i in range(len(palavra_jogada)):
        lista_espaços.append('_')    
    print(lista_espaços)

    print(lista_forca[0])#desenho da forca no início da jogada

    letra_incorreta = [] #lista de letras escolhidas incorretas
    letra_correta = [] #lista de letras escolhidas corretas
    letras_escolhidas = [] #lista de todas as letras escolhidas
    while len(letra_incorreta)< 7 and len(letra_correta) != len(lista_espaços):

        letra_escolhida = qual_letra() #letra que o jogador escolheu
        letras_escolhidas.append(letra_escolhida)
        
        if letra_escolhida in palavra_jogada:
            for i in range(len(palavra_jogada)):
                if letra_escolhida == palavra_jogada[i]:
                    lista_espaços[i] = letra_escolhida
                    letra_correta.append(letra_escolhida)
            print('Letras escolhidas que não estão na palavra da jogada: ', letra_incorreta)
            print(lista_espaços) 

        elif letra_escolhida not in palavra_jogada:
            letra_incorreta.append(letra_escolhida)
            print('Letras escolhidas que não estão na palavra da jogada: ', letra_incorreta)
            print(lista_espaços)

        print(lista_forca[len(letra_incorreta)])
        
    if len(letra_incorreta) == 7:
        print('Jogada perdida!')
        jogadas_perdidas += 1
        
    else:
        print('Jogada ganha!')
        jogadas_ganhas += 1
        
    print('Total de jogadas ganhas:', jogadas_ganhas)
    print('Total de jogadas perdidas:', jogadas_perdidas)


    quer_jogar = False
    
    while quer_jogar == False:
        jogada = str(input('Você quer jogar novamente (s/n)?')).upper()
        
        if not jogada == 'S' and not jogada == 'N':
            print("Escolha entre as opções s ou n")
        elif jogada == 'S': 
            quer_jogar = True
        elif jogada == 'N':
            quer_jogar = True          

    
```
